```python
#joining arrays
A1=np.array([34,56,78,90]).reshape(2,2)
B1=np.array([7,9,3,12]).reshape(2,2)
print(A1)
print("********")
print(B1)
print("********")
print("After vertically satcking arrays")#np.stack(((A1,B1),axis=0))
print(np.vstack((A1,B1)))
```

```python
#joining arrays
A1=np.array([34,56,78,90]).reshape(2,2)
B1=np.array([7,9,3,12]).reshape(2,2)
print(A1)
print("********")
print(B1)
print("********")
print("After horizontally stacking arrays")#np.stack(((A1,B1),axis=1))
print(np.hstack((A1,B1)))
```

```python
A=np.arange(30).reshape(2,3,5)
print(A)
print("After dstack")
print(np.dstack(A)) #no.of rows=no.of groups
#no.of columns in every group=no.of rows every group    i.e. (3,5,2)
#o/p will be [0,0],[1,1],......
            #[index,index]
```

```python
#Splitting arrays!
A=np.arange(18).reshape(6,3)
print(A)
print("after splitting")
np.split(A,(3))#(A,no.of equal array divisions)
```
