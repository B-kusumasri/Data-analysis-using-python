```python
#Hypothesis testing:
import numpy as np
from scipy import stats
```

```python
#two groups of data
data1=np.array([28,21,26,29,23])
data2=np.array([21,27,25,28,19])
print(data1.mean())
print(data2.mean())
```

```python
#perform t-test assuming unequal variances
t_stat,p_value=stats.ttest_ind(data1,data2,equal_var=False)
```

```python
#print the results
print("T-statistic",t_stat)
print("P-statistic",p_value)
```

```python
#set significance level
alpha = 0.05
```

```python
#Interpret the values
if p_value<alpha:
  print("Reject the null hypothesis.this means they are significant")
else:
  print("Reject the alternate hypothesis. this means they are not significant")
```
