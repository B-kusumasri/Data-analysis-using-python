```python
#there are 5 subjects phsycology,nano technology,medicine,oncology,anthropology.
#2 students have applied to da masters in any of these 5 fields with a line plot show in which field they can excel.The cut of for each subject is 75
stu1=[67,45,73,71,57]
stu2=[43,67,68,53,85]
cut=[70,70,70,70,70]
sub=['phsycology','nano technology','medicine','oncology','anthropology']
#student1
#syntax: plottype(X,Y,label,color,marker,linewidth,linestyle)
plt.plot(sub,stu1,label="student1")
#student2
plt.plot(sub,stu2,label="student2")
plt.plot(sub,cut,label="cutoff")
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.legend()
```

```python
stu1=[67,45,73,71,57]
stu2=[43,67,68,53,85]
cut=[70,70,70,70,70]
sub=['phsycology','nano technology','medicine','oncology','anthropology']
#student1
#syntax: plottype(X,Y,label,color,marker,linewidth,linestyle)
plt.scatter(sub,stu1,label="student1",color="y",marker="P") #P=Plus
#student2
plt.scatter(sub,stu2,label="student2",color='r',marker="H") #H=hexagon
plt.plot(sub,cut,label="cutoff")
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.legend()
```

```python
#An india vs australia match is on live. India has given 298 as target for australia. The runs for every file overs is given as follows
#I=[25,51,84,131,160,189,220,250,267,297]
#Austalia have completed 25 overs
#A=[15,41,94,110,151]
#Plot over vs runs for both countries with india as blue line and aus as yellow line with required and legends.The marker for india is Flag(>)
#and the marker for australia is +. Takes overs from a range function create runs as array.line width is 2.

from numpy import random as rd
I=np.array([25,51,84,131,160,189,220,250,267,297])
A=np.array([15,41,94,110,151])
IO=np.arange(5,51,5)
AO=np.arange(5,26,5)
plt.plot(IO,I,color="blue",label="India",marker=">",linewidth=2)
plt.plot(AO,A,color="yellow",label="Australia",marker="+",linewidth=2)
plt.title("India vs Aus")
plt.xlabel("overs")
plt.ylabel("runs")
plt.legend()
plt.grid()
```
