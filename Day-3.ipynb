Accesing the data from data frame

```python
#access the data
#loc=integer location, index
#loc=field names, index
#dfname.loc[index]----->rows
#dfname.loc[st:stop]----->range of rows
#dfname.loc[r_ind,col_name]----->rows and columns
mydf=pd.read_csv("/content/grades_withnulls.csv")
mydf
```

```python
mydf.loc[0]  #gives the first record
```

```python
mydf.loc[0:5]  #gives the records upto 5
```

```python
mydf.loc[5:7,"SEM1"]  #first record
```

iloc accepts only index numbers

```python
mydf.iloc[10:13,"SEM1"] #iloc does not takes the colnames
```

```python
mydf.iloc[10:13,0:4]
```

searching the data

```python
mydf[mydf.SEM1==9.8]
```

```python
#print records whose score is 9.3 in sem3
mydf[mydf.SEM3==9.3]
```

```python
mydf.loc[mydf.SEM1>9.8,'Names']    #loc[index,colnames]
```

```python
#print the grades of students who scored more than 9.7 in sem3
mydf.loc[mydf.SEM3>9.7,'Grade']
#accessing the records with sem3>9.7 but only printing grade
```
