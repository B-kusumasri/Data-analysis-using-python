ANOVA(one-way

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/Vaccine.csv")
df
```

```python
#Hypothesis testing:
import numpy as np
from scipy import stats
```

```python
a=df['Efficiency'].mean()
b=df['Rec_Rate'].mean()
print("Mean Recovery",b)
group_A=df['Efficiency']
group_B=df['Rec_Rate']
t_stat,p_value=stats.ttest_ind(group_A,group_B)
print("T-statistic",t_stat)
print("P-statistic",p_value)
```

```python
#set significance level
alpha=0.05
```

```python
#Interpret the values
if p_value<alpha:
  print("H0- related")
else:
  print("H1- not related")
```

ANOVA (one-way ANOVA)


```python
dfA=df[df['Vaccine']=='A']['Efficiency']
dfB=df[df['Vaccine']=='B']['Efficiency']
dfA
```

```python
f_stat,p_value=stats.f_oneway(dfA,dfB)

print("F-statistic:",f_stat)
print("p-value:",p_value)
```

```python
#Interpret the values
if p_value<alpha:
  print("H0- related")
else:
  print("H1- not related")
```
